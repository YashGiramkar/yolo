```python
import sys
sys.path.append('/content/drive/MyDrive/yolo')

from yolov1tiny import *
import random
import numpy as np

from google.colab.patches import cv2_imshow

random.seed(555)
```

```python
trainFiles, valFiles = getTrainVal()
jointFiles = trainFiles + valFiles
random.shuffle(jointFiles)

num_files = len(jointFiles)
num_train = int(np.floor(0.75 * num_files))

trainFiles = jointFiles[ : num_train]
valFiles = jointFiles[num_train : ]
```

```python
trainGen = getVOCData(trainFiles)
valGen = getVOCData(valFiles)
```

```python
def class_scheduler(epoch, lr):
    if epoch <= 10:
        return 0.001
    elif epoch > 10  and epoch <= 20:
        return 0.005
    elif epoch > 20 and epoch <= 30:
        return 0.01
    elif epoch > 20 and epoch <= 30:
        return 0.005
    else:
        return 0.001

classModel = yoloClassModel()

opt = tf.optimizers.SGD(learning_rate=0.0001, momentum=0.9)
lr_callback = tf.keras.callbacks.LearningRateScheduler(class_scheduler)
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=os.path.join('/content/drive/MyDrive/yolo_weights', 'classModel.{epoch:02d}-{loss:.2f}.h5'), \
                                                 save_weights_only=True, verbose=1, save_freq= len(trainFiles) // BATCHSIZE * 10)

classModel.compile(loss=yoloClassLoss, optimizer=opt)
```

```python
classModel.fit(x=trainGen,
        steps_per_epoch = len(trainFiles) // BATCHSIZE,
        epochs = 50,
        verbose = 2,
        validation_data = valGen,
        validation_steps = len(valFiles) // BATCHSIZE,
        callbacks = [lr_callback, cp_callback],
        )
```

```python
def printClassification(label, from_logits=False):
    for ridx, row in enumerate(label):
        for cidx, cell in enumerate(row):
            if from_logits:
                cell = tf.math.sigmoid(cell[:C])
            #print(cell)
            max_idx = tf.argmax(cell, axis=-1)
            print(PASCAL_VOC_CLASSES[max_idx], cell[max_idx])
```

```python
def calcClassAccuracy(yTrue, yPred):
    yTrue = K.reshape(yTrue, (-1, S, S, C + len_boxData))
    objProb = yTrue[..., C + 4 : C + 5]

    yPred = K.reshape(yPred, (-1, S, S, C))
    yPred = yPred[..., : C]

    classProb = yTrue[..., : C]
    pred_classProb = yPred[..., : C]

    pred_classProb = tf.math.sigmoid(pred_classProb)

    classifier_score = []
    obj_idx = np.argwhere(objProb == 1)
    for idx in obj_idx:
        idxClass = np.argmax(classProb[tuple(idx[:-1])])
        #print(classProb[tuple(idx[:-1])])
        pred_idxClass = np.argmax(pred_classProb[tuple(idx[:-1])])
        #print(pred_classProb[tuple(idx[:-1])])
        pred_val = pred_classProb[tuple(idx[:-1])][idxClass]
        if idxClass == pred_idxClass:
            classifier_score.append(1)
        else:
            classifier_score.append(0)
    
    return classifier_score
```

```python
score = calcClassAccuracy(labels, results)
np.sum(score)/len(score)
```

```python
def evalClassAccuracy(model, testGen):
    batch_accuracy = []
    for i in range(len(testGen)):
        images, labels = testGen[i]
        results = model(images)
        batch_score = calcClassAccuracy(labels, results)
        batch_accuracy.append(np.sum(batch_score)/len(batch_score))
    
    return np.sum(batch_accuracy) / len(batch_accuracy)

```

```python
A = evalClassAccuracy(classModel, valGen)
print(A)
```
